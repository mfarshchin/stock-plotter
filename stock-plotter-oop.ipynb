## This notebook is to analyze and plot history of a single stock (oop)

```python
from class_info import *
```

### stock information

```python
# Stock(`stock name`, `purchase price`)
stock = Stock('QS', 40)
```

### Get history

```python
df_d = stock.getHistory(days=365*5, interval='1d')
df_d
```

### Get volume distribution

```python
df_d_vd = stock.Vol_Dist(df_d, bins=100, decay_halftime=1)
df_d_vd
```

### Calculate moving averages

```python
df_d_dailyMA = stock.mySimpleMA(df_d,spans=[50])
df_d_dailyMA
```

### Plot results

```python
p = stock.plot_price_vol(df_d, df_d_vd,)
p = stock.plot_add_purchasePrice(p, line_dash='dashed')
p = stock.plot_add_MA(df_d_dailyMA, p)
show(p)
```
